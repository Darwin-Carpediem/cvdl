```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'dollar.webp'  
brightness_factor = 50 
modify_brightness(image_path, brightness_factor)
```

```python
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
original_image=image
image2=image
image3=image
image4=image
```

```python
def modify_brightness(image_path, brightness_factor=50):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Error: Unable to load image.")
        return
```

```python
brightened_image = cv2.add(image, brightness_factor)
```

```python
 darkened_image = cv2.subtract(image, brightness_factor)
```

```python
    plt.figure(figsize=(12, 8))
    plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
    plt.axis('off')
    plt.subplot(1, 3, 2), plt.imshow(brightened_image, cmap='gray'), plt.title('Brightened Image')
    plt.axis('off')
    plt.subplot(1, 3, 3), plt.imshow(darkened_image, cmap='gray'), plt.title('Darkened Image')
    plt.axis('off')
    plt.show()
```

```python
#2
alpha = 1.3 
beta = 0     
linear_scaled_image = cv2.convertScaleAbs(image2, alpha=alpha, beta=beta)
```

```python
    plt.figure(figsize=(16, 8))
    plt.subplot(1, 5, 1)
    plt.imshow(original_image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.subplot(1, 5, 4)
    plt.imshow(linear_scaled_image, cmap='gray')
    plt.title('Linear Scaling (Alpha=1.5)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
#3
kernel_size=5
box_filtered = cv2.blur(image2, (kernel_size, kernel_size))
```

```python
 gaussian_filtered = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigmaX=0)
```

```python
    plt.subplot(1, 3, 2)
    plt.imshow(box_filtered, cmap='gray')
    plt.title(f'Box Filter (Kernel={kernel_size}x{kernel_size})')
    plt.axis('off')
```

```python
    plt.subplot(1, 3, 3)
    plt.imshow(gaussian_filtered, cmap='gray')
    plt.title(f'Gaussian Filter (Kernel={kernel_size}x{kernel_size})')
    plt.axis('off')

    
```

```python
#4
median_filtered_image = cv2.medianBlur(image3, kernel_size)
```

```python
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image3, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(median_filtered_image, cmap='gray')
    plt.title(f'Median Filtered (Kernel={kernel_size}x{kernel_size})')
    plt.axis('off')
```

```python
#5
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
```

```python
 magnitude = cv2.magnitude(sobel_x, sobel_y)
```

```python
magnitude = cv2.convertScaleAbs(magnitude)
```

```python
    plt.figure(figsize=(10, 8))
    
    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(sobel_x, cmap='gray')
    plt.title('Sobel X (Horizontal Edges)')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(magnitude, cmap='gray')
    plt.title('Edge Magnitude (Sobel Edge Detection)')
    plt.axis('off')
```

```python
#6
threshold_value=100
thresholded_edges = cv2.threshold(magnitude, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(magnitude, cmap='gray')
    plt.title('Edge Magnitude (Sobel Edge Detection)')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(thresholded_edges, cmap='gray')
    plt.title(f'Thresholded Edges (Threshold={threshold_value})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
